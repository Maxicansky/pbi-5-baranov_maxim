# Задание 1

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
def classes (verdict):
    """ Функция для классификации фильмов по укрупненному рейтингу """
    if verdict['rating'] > 4:
        return "высокий рейтинг"
    elif verdict['rating'] > 2:
        return "средний рейтинг"
    else:
        return "низкий рейтинг"

ratings['class'] = ratings.apply(classes, axis = 1)   
ratings
```

# Задание 3

```python
movies = pd.read_csv('ml-latest-small\movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ml-latest-small\\ratings.csv')
ratings.head()
```

```python
years = range (1950, 2011)
type(years)
```

```python
def production_year (movies):
    years = range (1950, 2020)
    for year in years:
        if str(year) in movies['title']:
            return year
    return 1900
movies ['year'] = movies.apply(production_year, axis = 1)
movies.head()

    
```

```python
final_table = movies.merge(ratings, on='movieId', how='left')
final_table.head()
```

```python
final_table.groupby('year').mean()['rating'].reset_index().sort_values('rating', ascending=False).head()
```
