### Задание 1

Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа (решение на бумажке и на python).

```python
Image('Бин.jpg')
```

```python
simple_number = [2, 3, 5, 7, 11, 13]
simple_number
```

```python
p=0.5
n=15
rv = binom(n, p)
```

```python
frequency = [rv.pmf(simple_number[i]) for i in range(len(simple_number))]
```

```python
frequency
```

```python
sum(frequency).round(2)
```
