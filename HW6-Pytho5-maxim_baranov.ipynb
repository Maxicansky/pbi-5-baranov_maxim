# Первое задание
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0

```python
# Нам потребуются две таблицы ratings.csv и movies.csv
# Сначала работаем с ratings.csv
# Скачиваем в память как data_rating
data_rating = pd.read_csv('ratings.csv')
# Выделяем все строки с рейтингом 5
data_rating_five = data_rating [data_rating.rating == 5]
# Структурируем (когда шла отладка кода, так было удобнее производить анализ и контролировать промежуточные результаты)
data_rating_five = data_rating_five.sort_values(by = ['movieId'], ascending = [True])
# Группируем по Id фильма
data_rating_count = data_rating_five.groupby('movieId').count()
# Находим максимальное количество пятерок
max_number = data_rating_count['rating'].max()
# Выделяем фильм/фильмы с таким количеством пятерок 
data_rating_max = data_rating_count [data_rating_count.rating == max_number]
# Удаляем ненужные для конечного ответа столбцы (следовало сделать это раньше, для экономии памяти и ускорения работы)
data_rating_max = data_rating_max.filter(items = ['movieId', 'rating'])
data_rating_max
```

```python
# Формируем татафрейм на основе файла movies.csv
data_movies = pd.read_csv('movies.csv')
# Объединяем результаты, чтобы увидеть реквизиты фильма
final_table = data_rating_max.merge(data_movies, on='movieId', how='left')
# Приводим в читабельный вид и выводим на экран
print(final_table.rename(columns={'rating': 'quantity_of_five'}))
```

# Ответ: "Побег из Шоушенка"


# Второе задание
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
Не учитывайте в расчетах отрицательные значения quantity

```python
# Грузим файл. Смотрим, что получилось
data_power = pd.read_csv('power.csv')
data_power.head()
```

```python
# Последовательно фильтруем, по названию, по категории, по годам
data_rating_country = data_power [(data_power.country == 'Latvia') | (data_power.country == 'Lithuania') | (data_power.country == 'Estonia')]
data_rating_country_cat = data_rating_country [(data_rating_country.category == 4) | (data_rating_country.category == 12) | (data_rating_country.category == 21)]
data_rating_cou_cat_year = data_rating_country_cat [(data_rating_country_cat.year >= 2005) & (data_rating_country_cat.year <= 2010)]
data_rating_cou_cat_year
```

```python
# Здесь, искали Литву
data_power [data_power['country'].str.contains('Lit', case=False)]
```

```python
# Здесь, искали Эстонию
data_power [data_power['country'].str.contains('est', case=False)]
```

```python
# Суммируем и выводим ответ
summary = data_rating_cou_cat_year['quantity'].sum()
summary
```
