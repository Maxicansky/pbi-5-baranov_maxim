Построить частотный график для двух параметров: Freedom и positive affect. Какое это распределение?

**При построении частотного графика, для определения размера бина воспользуемся формулой Стерджесса**

```python
bin = int(round(math.log2(len(df['Freedom']))))
quantity = int(round(len(df['Freedom'])/bin))
title = 'Частотный график. Число бинов ' + str(quantity)
plt.hist(df['Freedom'], bin, facecolor='g')
plt.xlabel('frequency')
plt.ylabel('Freedom')
plt.title(title)
plt.grid(True)
plt.show()
```

```python
bin = int(round(math.log2(len(df['Positive affect']))))
quantity = int(round(len(df['Positive affect'])/bin))
title = 'Частотный график. Число бинов ' + str(quantity)
plt.hist(df['Positive affect'], int(round(math.log2(len(df['Positive affect'])))), facecolor='g')
plt.xlabel('frequency')
plt.ylabel('Positive affect')
plt.title(title)
plt.grid(True)
plt.show()
```

**Скорее всего мы имеем дело с равномерным распределением. Незначительный экстремум в правой части частотных графиков,   
вероятнее всего является недостатком выбора ширины бина.  
Мы уверено можем предположить, что каждой стране соответствует свое число в промежутке от 1 до 155.  
Скорее всего мы имеем дело не с результатами измерений или оценками экспертов, а с ранжиром в возрастающем порядке.  
Первое место, второе место и т.д.  
Проверим это путем сортировки таблицы**

```python
 df.sort_values(by = ('Freedom'))
```

```python
 df.sort_values(by = ('Positive affect'))
```

**Наше предположение оказалось верным - в обоих случаях мы имеем дело с ранжиром**

Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
print('Среднее значение равно ', statistics.mean(df['Freedom']))
print('Медиана равна ', statistics.median(df['Freedom']))
try:
    print('Мода равна ', statistics.mode(df['Freedom']))
except:
    print('Нет уникальной моды, все 155 значений встречаются одинаковое количество раз')
print('Минимальное значение равно ', min(df['Freedom']))
print('Максимальное значение равно ', max(df['Freedom']))
print('Стандартное отклонение равно ', statistics.pstdev(df['Freedom']))
print('Дисперсия равна ', statistics.pvariance(df['Freedom']))
print('Размах равен ', max(df['Freedom'])-min(df['Freedom']))
print('Был один выброс, но мы удалили его ранее')
```

**Существуют аналитические решения для равномерного распределения**

```python
print('Среднее значение равно ', (len(df['Freedom'])+1)/2)
print('Медиана равна ', (len(df['Freedom'])+1)/2)
print('У равномерного распределения отсутствует мода')
print('Минимальное значение равно ', min(df['Freedom']))
print('Максимальное значение равно ', max(df['Freedom']))
print('Стандартное отклонение равно ', ((len(df['Freedom'])*len(df['Freedom'])-1)/12)**0.5)
print('Дисперсия равна ', (len(df['Freedom'])*len(df['Freedom'])-1)/12)
print('Размах равен ', max(df['Freedom'])-min(df['Freedom']))
print('Был один выброс, но мы удалили его ранее')
```

**Полное совпадение результатов указывают на то, что изучаемое распределение является равномерным**

Найти корреляцию в следующих колонках:

Корреляция между: 4046 - Total number of avocados with PLU 4046 sold  и например Average Price. Если корреляция есть, то скорее всего, данный вид авокадо дорогой по цене и он влияет на среднюю стоимость.

**Корреляция слабая и к тому же отрицательная, первоначальная гипотеза не верна.

```python
df_6 = df[df['4046'] > 6000000]
df_6.head()
```

```python
corr, p_value  = pearsonr(df_6['4046'], df_6['AveragePrice'])
corr
```

```python
sns.jointplot(x='4046', y='AveragePrice', 
              data=df_6, kind='reg')
```

**В те дни, когда продавалось много авокадо данного типа наблюдалась умеренная корреляционная    
связь между количеством проданного авокадо '4046' и средней ценой.  
Удивительно, что коэффициент корреляции отрицательный. Можно выдвинуть гипотезу,
