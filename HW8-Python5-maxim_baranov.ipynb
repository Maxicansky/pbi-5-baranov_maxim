# Задание 1
Посчитайте среднее время жизни пользователей, 
которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным 
и минимальным значением столбца timestamp для данного значения userId

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
group_by_userId = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'rating': 'sum', 'movieId': 'count'})
group_by_userId.head()
```

```python
group_by_userId_filtered = group_by_userId [group_by_userId.movieId['count'] > 100]
group_by_userId_filtered.head()
```

```python
group_by_userId_filtered['difference'] = group_by_userId_filtered.timestamp['max']-group_by_userId_filtered.timestamp['min']
group_by_userId_filtered.head()
```

```python
average_lifetime = group_by_userId_filtered['difference'].mean()
average_lifetime
```

# Задание 2

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:

- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air = air [['client_id', 'air_revenue']]
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)

client_base
```

```python
rzd_auto = pd.merge(rzd, auto, how='outer', on='client_id')
rzd_auto
```

```python
rzd_auto_air = pd.merge(rzd_auto, air, how='outer', on='client_id')
rzd_auto_air
```

```python
rzd_auto_air_address = pd.merge(rzd_auto_air, client_base, how='outer', on='client_id')
rzd_auto_air_address
```
